### Scatter Plots

```python
# show relationship between two numeric values
```

```python
sat = pd.read_csv('datasets/scores.csv')
```

```python
sat.head()
```

```python
# is reading and writing more strongly associated than
# writing and math
```

```python
math = sat[sat['SAT Section']=='Math'].Score.reset_index(drop=True)
reading = sat[sat['SAT Section']=='Reading'].Score.reset_index(drop=True)
writing = sat[sat['SAT Section']=='Writing'].Score.reset_index(drop=True)
```

```python
scores = pd.concat([math,reading,writing],axis=1)
scores.columns = ['Math','Reading','Writing']
```

```python
scores.head()
```

```python
scores.loc[:,['Writing','Reading']].plot(kind='scatter',x='Reading',y='Writing');
scores['Reading'].corr(scores['Writing'])
```

```python
scores.loc[:,['Math','Reading']].plot(kind='scatter',x='Reading',y='Math');
scores['Reading'].corr(scores['Math'])
```

```python
scores.loc[:,['Math','Writing']].plot(kind='scatter',x='Writing',y='Math')
scores['Math'].corr(scores['Writing'])
```

```python
scores['Reading'].corr(scores['Writing'])
```

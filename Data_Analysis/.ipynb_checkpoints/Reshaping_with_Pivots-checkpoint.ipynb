### The Problem with Average Percentage

```python
scores.pivot_table(values='Percent Tested',index='Borough',aggfunc='mean')
```

```python
scores['Taker'] = scores['Percent Tested'] * scores['Student Enrollment'] / 100
```

```python
scores.head()
```

```python
scores.pivot_table(values='Taker', index='Borough', aggfunc='mean')
```

```python
scores.loc[scores.Borough=='Bronx']['Taker'].mean()
```

```python
scores.pivot_table(values='Student Enrollment', index='Borough', aggfunc='mean')
```

```python
# to get a real picture, we have to find average test taker scaled
# by student enrollment
```

```python
scores.groupby(['Borough'])['Student Enrollment','Taker']\
.apply(lambda sg: sg['Taker'].mean()/sg['Student Enrollment'].mean())\
.nsmallest()
```

```python
scores.groupby(['Borough'])['Student Enrollment','Taker']\
.apply(lambda sg: sg['Taker'].sum()/sg['Student Enrollment'].sum())\
.nsmallest()
```

### Applying Multiple Functions

```python
# show min and max score by borough in a single step (one pivot table)
```

```python
scores.pivot_table('Score','Borough','SAT Section', aggfunc=['max','min',np.mean,'std',np.min]) 
```

```python
scores.pivot_table('Score','Borough','SAT Section', aggfunc=['max','min',np.mean,'std',np.min]).rename(axis=1,level=0, mapper={'max':'Max','amin':'npmin'})
```

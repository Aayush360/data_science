### Many-to-Many Joins

```python
# when we have duplicates in the key columns 
# from both the left and right object
```

```python
# creating a survey data

survey = pd.DataFrame({
    'School Type':['Ivy League','Ivy League','Engineering','Engineering'],
    'Prestige': ['High','Good','Good','Okay'],
    'Respondent': [1,2,3,4]
})
```

```python
survey
```

```python
ives.head()
```

```python
# merge field is school type
# duplicate in both dfs
```

```python
pd.merge(ives,survey)

# 8 (ives df) * 2 (in survey df)
```

```python
pd.merge(ives,survey).shape
```

```python
survey_new = survey.append(pd.Series({'School Type':'Ivy League',
                        'Prestige':'Very High',
                        'Respondent':5}, name=4))
```

```python
pd.merge(ives,survey_new)

# 8 *3 = 24
# cartesian product
```

```python
import pandas as pd
import numpy as np
```

```python
names = ['Aayush','Andrew','Nicole']
```

```python
ages = [20,20,21]
```

```python
married = [True, False, True]
```

```python
# dict of tuples
```

```python
tuple_names = tuple(names)
tuple_ages = tuple(ages)
tuple_married = tuple(married)
```

```python
pd.DataFrame({'name':tuple_names,
             'ages':tuple_ages,
             'married':tuple_married})
```

### dict of series

```python
series_names = pd.Series(names)
series_ages = pd.Series(ages)
series_married = pd.Series(married)
```

```python
pd.DataFrame({'name':series_names,
             'ages':series_ages,
             'married':series_married})
```

```python
pd.DataFrame({'names':['Olga','Andrew']})
```

```python
pd.DataFrame({'names':{0:'Olga',1:'Andrew'}})
```

### the enumerate function

```python
enumerate(names)
```

```python
list(enumerate(names))
# gives pairs of tuples containing value of each item
# as well as its poition
```

```python
# dict of dicts
```

```python
name_dict = {'name':{}}
```

```python
for i in list(enumerate(names)):
    name_dict['name'][i[0]]=i[1]
```

```python
name_dict
```

```python
dict_name = {k:v for k,v in enumerate(names)}
dict_ages = {k:v for k,v in enumerate(ages)}
dict_married = {k:v for k,v in enumerate(married)}
```

```python
dict_married
```

```python
def list_to_dict(l1):
    return {k:v for k,v in enumerate(l1)}
```

```python
list_to_dict(names)
```

```python
pd.DataFrame({'name':dict_name,
             'ages':dict_ages,
             'married':dict_married})
```

```python
# list of dicts
```

```python
pd.DataFrame([{'name':'Olga',
              'ages':29,
              'married':True}])
```

## Sampling with Replacement or Weights

```python
nutrition.sample(n=3,replace=True) 
# there can be changes of getting the same row multiple time althougth it is highly unlikely
```

```python
# weighted sampling
```

```python
weights = pd.Series(data=[10,10,10,1,2],index=[7,17,29,5,6])
```

```python
weights
```

```python
nutrition.sample(n=3,weights=weights)

# choosing 3 samples based on weights given to the indices
```

```python
# # the function should:
# - apply a random string capitalization method
# - to a sequence of value
# - return the transformed sequence
```

```python
def random_case(x):
    func = [x.str.swapcase, x.str.lower, x.str.title, x.str.upper]
    return choice(func)()
    
```

```python
soc.select_dtypes(include=object).transform(random_case)
```

### View vs Copy

```python
# copy : replica of what is in the dataframe
        # different distinct object , if we cahnge, doesnot reflect in the dataframe
# view : window into the underlying data (opening that gives direct access to the data) 
        # what we see is what is really thre in the dataframe
    # if we modify what we see, we end up modifying the dataframe
```

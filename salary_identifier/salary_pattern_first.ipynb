```python
def find_regular_income_pattern(df, end_date, income_segment_id,conf):
    """Identify salary like patterns.

    Arguments:
    ---------
    df: DataFrame, input data for salary like pattern identification.
    end_date: dict, dictionary consisting of year, month and day of latest transaction date.
    income_segment_id: int, Income segmentation Id.

    Returns:
    -------
    DataFrame.
    """
    try:
        config = conf

        df["bins"], bin_mapping = find_binning.get_binning(df.amount, income_segment_id)

        result = df.groupby(["account_number", "tran_date", "bins"]).agg(
            freq=("bins", "count"))
        result.reset_index(inplace=True)

        result = result.groupby(["account_number", "bins"]).agg(
            min_date=("tran_date", "min"), max_date=("tran_date", "max"),
            distinct_months=("tran_date", "nunique"),
            freq=("bins", "count"), total_freq=("freq", "sum")).reset_index()

        result["lower"] = result["bins"].map(bin_mapping)
        result["upper"] = result["lower"].apply(lambda x: x[1])
        result["lower"] = result["lower"].apply(lambda x: x[0])

        result["month_diff"] = result[["min_date", "max_date"]].apply(
            lambda x: (x[1].year - x[0].year) * 12 + (x[1].month - x[0].month) + 1, axis=1)

        result["recency"] = result["max_date"].apply(lambda x:
                                                     (end_date.year - x.year) * 12 + (end_date.month - x.month) + 1)

        # Filtering
        result = result.query("recency <= {} and distinct_months >= {}".format(
            3, 3)).copy()

        result.sort_values(by=["account_number", "recency", "distinct_months"],
                           ascending=(True, True, False), inplace=True)
        result.drop_duplicates(
            subset=["account_number"], keep="first", inplace=True)
        result = result[["account_number", "lower", "upper"]].copy()

        return result
    except Exception as err:
#         logger.error(err)
        raise err
```

### Binning algorithm

```python
def get_binning(series, income_segment_id,conf, max_val=None):
    """Return binned data.

    Arguments:
    ---------
    series: Series
        input data for binning.
    income_segment_id: int, Income Segmentation ID.
    max_val: int
        Max value of deposit transactions to identify salary.


    Returns:
    -------
    Series, dict
    """
    try:
        config = conf

        min_val = min(500, series.min())

        max_val = series.max() if max_val is None else max_val

        bin_range, labels = generate_bin_range(min_val, max_val, config)

        bin_range[0] = bin_range[0] - 0.001

        series_bins = pd.cut(series, bins=bin_range,
                             retbins=False, labels=labels)

        series_bins = series_bins.astype("int32")
        bin_range[0] = bin_range[0] + 0.001

        bin_ranges = []
        for i in range(len(bin_range) - 1):
            bin_ranges.append((bin_range[i], bin_range[i + 1]))

        bin_mapping = dict(zip(labels, bin_ranges))

        return series_bins, bin_mapping
    except Exception as err:
#         logger.error(err)
        raise err


def generate_bin_range(min_val, max_val, config):
    """Generate bin range.

    Arguments:
    --------
    min_val: int
        Min value of deposit transactions to identify salary.
    max_val: int
        Max value of deposit transactions to identify salary.
    config: dict, income segment specific configuration.

    Returns:
    --------
    array, list

    """
    try:
        bin_amount = 20000
        breaker = 10
        bin_amount_rate = 15000
        bins_arr = np.array([])

        while True:
            if bins_arr.size == 0:
                bins_arr = np.hstack(
                    (bins_arr, (np.ones(breaker) * bin_amount).cumsum()))
            else:
                bins_arr = np.hstack((bins_arr, np.hstack(
                    (np.array([bins_arr[-1]]), (np.ones(breaker) * bin_amount))).cumsum()[1:]))

            bin_amount += bin_amount_rate
            if bins_arr[-1] > max_val:
                break

        bins_arr = bins_arr + 10000
        bins_arr = np.hstack((np.array([min_val]), bins_arr))
        labels = list(range(1, bins_arr.size))

        return bins_arr, labels
    except Exception as err:
#         logger.error(err)
        raise err
```
